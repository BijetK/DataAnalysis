## 단계구분도

```python
import pandas as pd
import folium
import json
```

```python
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
with open(filename) as json_file:
    geo_data = json.load(json_file)
```

- 자치구별 공원 면적

```python
center = [37.581, 126.986]      # 서울 중심부 위도, 경도
map = folium.Map(location = center, zoom_start = 11, tiles = 'Stamen Toner')

folium.Choropleth(
    geo_data = geo_data,
    data = df,
    columns = ('지역', '공원면적'),
    key_on = 'feature.id',
    fill_color = 'BuPu',
    legend_name = '공원면적'
).add_to(map)
title = '<h3 align="center" style="font-size:20px">서울시 자치구 별 공원면적</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
park = pd.read_csv(filename)
park.head()
```

```python
map = folium.Map(location = center, zoom_start = 11, tiles = 'Stamen Toner')

folium.Choropleth(
    geo_data = geo_data,
    data = df,
    columns = ('지역', '공원면적'),
    key_on = 'feature.id',
    fill_color = 'BrBG',
    legend_name = '공원면적'
).add_to(map)
for i in park.index:
    folium.CircleMarker([park.위도[i], park.경도[i]], 
                        radius=int(park.크기[i]),
                        tooltip=f"{park.공원명[i]}({park.면적[i]:,d}㎡)",
                        color='#199c6e', fill_color='#199c6e').add_to(map)
title = '<h3 align="center" style="font-size:20px">서울시 자치구 별 공원면적</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
