# 반복문으로 전체

```python
list = []
for file in glob(f'{data_folder}/*.csv'):
    list.append(file)

list
```

```python
df_korea = pd.DataFrame()

for filename in list:
    df = pd.read_csv(f'{filename}', encoding='utf-8')

    # 필요한 행만 남기기
    df = df[['상권업종중분류코드', '상호명','지점명','시도명','시군구명','도로명주소']]

    # Q12만 남기기
    df=df[df.상권업종중분류코드 == 'Q12']

    # 커피샵 상호명에 따라 분류
    
    # df_star = df[df.상호명.str.contains('스타벅스|STARBUCKS|스타박스', case = Fasle)]
    df_star=df[df.상호명.str.contains('스타벅스') | df.상호명.str.contains('STARBUCKS') | df.상호명.str.contains('starbucks') | df.상호명.str.contains('Starbucks')]
    df_idya=df[df.상호명.str.contains('이디야') | df.상호명.str.contains('EDIYA')| df.상호명.str.contains('ediya')| df.상호명.str.contains('Ediya')]
    df_cfbn=df[df.상호명.str.contains('커피빈') | df.상호명.str.contains('COFFEEBEAN') | df.상호명.str.contains('coffeebean')| df.상호명.str.contains('Coffeebean')]
    df_back=df[df.상호명.str.contains('빽다방') | df.상호명.str.contains('PAIKS') | df.상호명.str.contains('paiks')| df.상호명.str.contains('Paiks')]
    df_star['커피샵'] = '스타벅스'
    df_idya['커피샵'] = '이디야'
    df_cfbn['커피샵'] = '커피빈'
    df_back['커피샵'] = '빽다방'

    # 합치기
    df_all = pd.concat([df_star,df_idya,df_cfbn,df_back])

    # ID 따기 편하게 칼럼명 바꾸기
    df_all.columns = ['상권업종중분류코드', '상호명', '지점명', '광역시도', '시군구', '도로명주소', '커피샵']

    # 전국 한번에 합치기
    df_korea = pd.concat([df_korea, df_all])
```
