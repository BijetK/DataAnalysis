```python
df = df[['총계', '2017년 이전', '2018년', '2019년','2020년','2021년']]
df.head()
```

## 인구수 데이터 처리하기

```python
df = pd.read_csv(f'{data_folder}/서울시주민등록인구.csv', encoding='euc-kr')
del df['Column1']
df.head(10)
```

```python
tdf = df.T
tdf.reset_index(inplace=True)
del tdf[0]
del tdf[1]
tdf.set_index(tdf[2], inplace=True)
del tdf['index']
del tdf[2]
tdf.head()
```

```python
df = tdf.T
df.reset_index(inplace=True)
del df['index']
df.set_index(df['자치구'], inplace=True)
del df['자치구']
```

```python
df.columns = ['세대', '계', '남자', '여자','한국인계', '한국인남자', '한국인여자','외국인계','외국인남자','외국인여자','세대당인구','65세이상고령자']
df.head()
```

```python
def str2int(x):
    return int(str(x).replace(',',''))
```

```python
df['세대'] = df['세대'].apply(str2int)
df['계'] = df['계'].apply(str2int)
df['남자'] = df['남자'].apply(str2int)
df['여자'] = df['여자'].apply(str2int)
df['한국인계'] = df['한국인계'].apply(str2int)
df['한국인남자'] = df['한국인남자'].apply(str2int)
df['한국인여자'] = df['한국인여자'].apply(str2int)
df['외국인계'] = df['외국인계'].apply(str2int)
df['외국인남자'] = df['외국인남자'].apply(str2int)
df['외국인여자'] = df['외국인여자'].apply(str2int)
df['65세이상고령자'] = df['65세이상고령자'].apply(str2int)
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.to_csv('/content/mnt/MyDrive/results/CCTV/서울시주민등록인구.csv', index = False)
```
