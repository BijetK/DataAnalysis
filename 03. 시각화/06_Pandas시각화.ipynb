```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family = 'NanumBarunGothic')
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

## Line Plot

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2021', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title('Pandas의 plot()매서드 사용 예')
plt.xlabel('일자')
plt.ylabel('Data')
plt.grid()
plt.show()
```

## Bar Chart

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)      #[:20] : 20개 까지의 Series
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0,                           #sort_values : 정렬, # barh : 수평 그래프
                            color = plt.get_cmap('RdBu')(np.linspace(0.8, 0.2, 20)))      
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")      # barh로 바꾸면 x,y label 바꿔줘야함
plt.xlabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot = 0)
plt.ylim(0,7)       # 범례가 그래프 가리지 않게.
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = 'feature'        # 범례 제목에 feature 들어감
df2
```

```python
df2.plot.bar(rot = 0)
plt.title('각 종의 feature별 평균)')
plt.ylim(0, 8)
plt.show()
```

```python
df2.T
```

```python
df2.T.plot.bar(rot = 0)
plt.title('각 종의 feature별 평균)')
plt.ylim(0, 8)
plt.show()
```

## Pie Chart

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct = '%.2f%%')
plt.title('선실 별 승객 수 비율')
plt.axis('equal')       # 파이차트는 필수!!!
plt.show()
```

## Histogram

```python
iris.plot.hist()
plt.title("각 Feature 값들의 빈도수 Histogram")
plt.xlabel("데이터 값")
plt.show()
```

## Box plot

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by = 'species')        # by : 그룹핑을 해줌
plt.tight_layout(pad = 2, h_pad = 1)
plt.title('XXXX')       # 전체 title이 아닌 마지막 녀석만 이름이 바뀜 
plt.show()
```
